```python
import numpy as np
import pandas as pd
```

```python
dat = pd.read_csv('data/IT_tickets.csv')
```

```python
dat
```

```python
dat.shape
```

```python
# 2 is the automated ticketing system. 1 and 3 are the it desk and customer respectively.  
dat_filtered = dat[dat['article_sender_type_id'] != 2]
```

```python
dat_filtered
```

```python
dat_filtered.shape
```

```python
print('{} automated ticketing system articles removed.'.format(dat.shape[0] - dat_filtered.shape[0]))
```

```python
dat = dat_filtered
```

```python
dat['a_subject'].value_counts()
```

```python
# Remove all rows with subject "Close!" as it indicates a ticket closing, "Note!" as is a note, "Responsible Update!", 
# "Priority Update!", "Change Queue!", "Owner Update!", "Ticket Merged", "Pending!", "Phone Call!", 
# "Undelivered Mail Returned to Sender", "MailCleaner quarantine report"

subjects_to_purge = ["Close!", "Note!", "Responsible Update!", "Priority Update!", "Change Queue!", "Owner Update!", 
                     "Ticket Merged", "Pending!", "Phone Call!","Undelivered Mail Returned to Sender", 
                     "MailCleaner quarantine report"]

dat_filtered = dat[~dat['a_subject'].isin(subjects_to_purge)]
```

```python
print('{} automated ticketing system articles removed.'.format(dat.shape[0] - dat_filtered.shape[0]))
print('New row count {}'.format(dat_filtered.shape[0]))
dat = dat_filtered
```
